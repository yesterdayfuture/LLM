```python
# 配置硬件参数
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
#配置量化参数
bnb_config = BitsAndBytesConfig(
    load_in_4bit= True,
    bnb_4bit_quant_type= "nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant= True
)
```

```python
#模型进行量化
model = AutoModelForCausalLM.from_pretrained(
    model_dir,
    quantization_config= bnb_config,
    device_map= "auto",
    trust_remote_code= True
)
```

```python
#配置LoRA参数
lora_config = LoraConfig(
    r=64,                           # 低秩矩阵维度（平衡效果与速度）[4](@ref)
    lora_alpha=128,                 # 缩放系数（建议设为r的2倍）[2](@ref)
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj"], # 目标注意力模块[5](@ref)
    lora_dropout=0.05,              # 防止过拟合的Dropout率[4](@ref)
    bias="none",                    # 不训练偏置参数[2](@ref)
    task_type="CAUSAL_LM"           # 任务类型（因果语言模型）[4](@ref)
)
```

```python
#获取量化模型
model = prepare_model_for_kbit_training(model)

#模型应用 微调参数
model = get_peft_model(model, lora_config)

#打印可训练参数占比
model.print_trainable_parameters()
```

```python
#加载数据集
# 步骤3：加载并处理数据集
dataset = load_dataset("yahma/alpaca-cleaned")  # 使用HuggingFace指令数据集[5](@ref)

def format_instruction(sample):
    # 将数据集转换为Alpaca指令格式[2](@ref)
    return {
        "text": f"""Below is an instruction that describes a task. Write a response that appropriately completes the request.

### Instruction:
{sample['instruction']}

### Input:
{sample['input']}

### Response:
{sample['output']}"""
    }

dataset = dataset.map(format_instruction)
tokenized_dataset = dataset.map(
    lambda x: tokenizer(
        x["text"],
        padding="max_length",
        truncation=True,
        max_length=512,
        return_tensors="pt"
    ),
    batched=True
)
```

```python
#配置训练参数
#输出的位置
peft_out = "./lianghua"

train_config = TrainingArguments(
    peft_out,
    per_device_train_batch_size=2,   # 根据显存调整（RTX3090建议2-4）[2](@ref)
    gradient_accumulation_steps=4,   # 梯度累积步数（等效batch_size=8）[5](@ref)
    num_train_epochs=20,              # 训练轮次（指令微调通常1-3轮）[4](@ref)
    learning_rate=2e-5,              # 初始学习率（QLoRA建议2e-5~3e-5）[1](@ref)
    fp16=True,                       # 混合精度训练（A100可用bf16）[5](@ref)
    logging_steps=10,                # 每10步记录日志[4](@ref)
    optim="paged_adamw_32bit",       # 分页优化器防止显存溢出[5](@ref)
    report_to="tensorboard",         # 使用TensorBoard监控训练
    save_strategy="epoch",           # 每轮保存检查点[4](@ref)

)
```

```python
#创建Trainer并启动训练
trainer = Trainer(
    model=model,
    args=train_config,
    train_dataset=tokenized_dataset['train'][:40000],
    eval_dataset=tokenized_dataset['train'][40000:],
    data_collator=lambda data: {
        "input_ids": torch.stack([f["input_ids"][0] for f in data]),
        "attention_mask": torch.stack([f["attention_mask"][0] for f in data]),
        "labels": torch.stack([f["input_ids"][0] for f in data])
    }
)
```

## transformers 对大模型进行sft（监督微调）

```python
import torch
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer
)
import evaluate
import numpy as np
```

```python
##############################
# 1. 数据集更换为IMDB影评数据集（二分类）
##############################
dataset = load_dataset("imdb")  # 无需API key的公开数据集

# 划分验证集（原数据集已自带测试集）
split_dataset = dataset["train"].train_test_split(
    test_size=0.2,
    seed=42
)
train_dataset = split_dataset["train"]
eval_dataset = split_dataset["test"]

##############################
# 2. 文本编码处理（适配IMDB文本字段）
##############################
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

def tokenize_function(examples):
    return tokenizer(
        examples["text"],  # IMDB的文本字段名也是"text"
        padding="max_length",
        truncation=True,
        max_length=256     # 缩短序列长度加速训练
    )

tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_eval = eval_dataset.map(tokenize_function, batched=True)
```

```python
##############################
# 3. 模型初始化（网页1的全参数微调策略）
##############################
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-cased",
    num_labels=2  # 情感标签0-4对应1-5星
)

# 确保全参数可训练（默认状态，无需冻结层）
for param in model.parameters():
    param.requires_grad = True
```

```python
print(model)
```

```python
##############################
# 4. 训练参数配置（综合网页2的超参数建议）
##############################
training_args = TrainingArguments(
    report_to="none",  # ✨ 关键修改：禁用所有日志服务
    output_dir="./imdb_finetune_output",
    evaluation_strategy="epoch",
    per_device_train_batch_size=16,  # 减小序列长度后可增大batch_size
    per_device_eval_batch_size=32,
    num_train_epochs=5,             # 减少训练轮次
    learning_rate=3e-5,             # 比预训练更小的学习率
    weight_decay=0.01,                   # 权重衰减防过拟合
    warmup_ratio=0.1,                    # 预热步数比例
    logging_steps=50,                    # 每50步记录日志
    save_strategy="epoch",               # 每轮保存检查点
    load_best_model_at_end=True,         # 训练结束加载最佳模型
    metric_for_best_model="accuracy",    # 监控指标
    fp16=True,                           # 混合精度训练（节省显存）
    gradient_accumulation_steps=2,       # 梯度累积补偿batch_size不足
    max_grad_norm=1.0                    # 梯度裁剪（防梯度爆炸）
)
```

```python
##############################
# 5. 评估指标设置（网页3的准确率计算）
##############################
metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    """计算分类准确率"""
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

##############################
# 6. 训练器初始化与执行（网页1的SFT流程）
##############################
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_eval,
    compute_metrics=compute_metrics
)
```

```python
# import os
# os.environ["HF_HUB_DISABLE_IMPLICIT_TOKEN"] = "1"  # 禁用隐式登录请求

# 启动全参数微调
trainer.train()

##############################
# 7. 测试示例适配二分类
##############################
trainer.save_model("./imdb_finetuned_model")

# 修改测试文本和输出解释
test_samples = [
    "This movie is a masterpiece! The acting is brilliant.",
    "A waste of time. Poor plot and terrible dialogue."
]

for text in test_samples:
    inputs = tokenizer(text, return_tensors="pt").to("cuda")
    with torch.no_grad():
        outputs = model(**inputs)
    predicted_label = torch.argmax(outputs.logits).item()
    sentiment = "positive" if predicted_label == 1 else "negative"
    print(f"Text: {text}\nPredicted: {sentiment}\n{'-'*50}")
```

```python
# 加载示例数据集（使用IMDB评论作为prompt-response对）
dataset = load_dataset("imdb", split="train[:1%]")  # 取1%作为示例

print(dataset.column_names)

def format_sft_data(examples):
    # 假设用户评论作为prompt，正面评价作为理想response
    return {"text": f"Prompt: {examples['text']}\nResponse: {'Positive' if examples['label']==1 else ''}"}
dataset = dataset.map(format_sft_data, batched=False)

print(dataset[0])
```

```python
# 配置LoRA高效微调
peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

# 初始化模型（网页1的SFT实现）
model = AutoModelForCausalLM.from_pretrained(
    mk_dir,
    device_map="auto",
    torch_dtype=torch.bfloat16
)
# 监督微调训练参数（网页1的训练配置）
training_args = TrainingArguments(
    output_dir="./sft_output",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=1,
    learning_rate=2e-5,
    num_train_epochs=2,
    fp16=True,
    max_steps=100,  # 示例用少量步数
    report_to="none"
)

# 创建SFT训练器（网页7的SFTTrainer使用）
sft_trainer = SFTTrainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    peft_config=peft_config
)

sft_trainer.train()
model.save_pretrained("./sft_model")
```

```python
##############################
# 阶段2：奖励模型训练
##############################
class RewardModelTrainer:
    def __init__(self):
        self.tokenizer = AutoTokenizer.from_pretrained(mk_dir)

        # 修正1：使用自定义模型结构
        class CustomRewardModel(torch.nn.Module):
            def __init__(self, base_model):
                super().__init__()
                self.model = base_model
                self.loss_fn = torch.nn.MarginRankingLoss(margin=1.0)

            def forward(self, chosen_inputs, rejected_inputs):
                # 正样本的奖励值
                chosen_rewards = self.model(**chosen_inputs).logits
                # 负样本的奖励值
                rejected_rewards = self.model(**rejected_inputs).logits
                # 计算对比损失
                loss = self.loss_fn(
                    chosen_rewards,
                    rejected_rewards,
                    torch.ones_like(chosen_rewards)
                )
                return {"loss": loss}

        # 加载基础模型
        base_model = AutoModelForSequenceClassification.from_pretrained(
            mk_dir,
            num_labels=1,
            problem_type="regression"
        )
        # 冻结参数
        for param in base_model.parameters():
            param.requires_grad = False
        base_model.score.weight.requires_grad = True

        self.model = CustomRewardModel(base_model)

    def train(self):
        # 构造数据集（修正为对比数据格式）
        def generate_dummy_data():
            return [
                {
                    "chosen": "Why did the chicken cross the road? To get to the other side!",
                    "rejected": "Chicken road crossing..."
                } for _ in range(10)
            ]

        # 数据预处理重构
        def tokenize_pair(example):
            # 正样本编码
            chosen = tokenizer(
                example["chosen"],
                padding="max_length",
                max_length=128,
                return_tensors="pt"
            )
            # 负样本编码
            rejected = tokenizer(
                example["rejected"],
                padding="max_length",
                max_length=128,
                return_tensors="pt"
            )

            return {
                "chosen_input_ids": chosen.input_ids[0],
                "chosen_attention_mask": chosen.attention_mask[0],
                "rejected_input_ids": rejected.input_ids[0],
                "rejected_attention_mask": rejected.attention_mask[0]
            }

        # 模型前向传播适配
        class CustomRewardModel(torch.nn.Module):
            def forward(self, chosen_input_ids, chosen_attention_mask,
                        rejected_input_ids, rejected_attention_mask):
                chosen_rewards = self.model(
                    input_ids=chosen_input_ids.unsqueeze(0),
                    attention_mask=chosen_attention_mask.unsqueeze(0)
                ).logits
                rejected_rewards = self.model(...).logits
                loss = self.loss_fn(...)
                return {"loss": loss}

        # 训练执行
        trainer = Trainer(
            model=model,
            args=training_args,
            train_dataset=tokenize_pair(generate_dummy_data())["train"].remove_columns(["chosen", "rejected"])
        )
        trainer.train()
        trainer.save_model("./reward_model")
```

```python
##############################
# 阶段3：PPO强化学习优化
##############################
# 结合网页2和网页5的PPO实现逻辑
def ppo_training():
    # 加载SFT模型和奖励模型（网页2的四模型架构）
    sft_model = AutoModelForCausalLMWithValueHead.from_pretrained(
        mk_dir,
        device_map="auto",
        torch_dtype=torch.bfloat16
    )
    # 加载修正后的奖励模型
    class RewardWrapper(torch.nn.Module):
        def __init__(self):
            super().__init__()
            self.model = AutoModelForSequenceClassification.from_pretrained("./reward_model")

        def forward(self, responses):
            inputs = tokenizer(responses, padding=True, return_tensors="pt").to("cuda")
            return self.model(**inputs).logits

    reward_model = RewardWrapper()

    # 初始化PPOTrainer（网页5的PPO配置）
    ppo_trainer = PPOTrainer(
        model=sft_model,
        config=TrainingArguments(
            per_device_train_batch_size=4,
            learning_rate=1e-6,
            output_dir="./ppo_output"
        ),
        tokenizer=tokenizer
    )

    # 生成训练数据（网页5的响应生成逻辑）
    def generate_prompts():
        return ["Explain quantum physics in simple terms", "Write a poem about AI"]

    # PPO训练循环（网页2的交互流程）
    for _ in range(10):  # 示例训练轮数
        prompts = generate_prompts()
        responses = []
        for prompt in prompts:
            inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
            outputs = sft_model.generate(**inputs, max_new_tokens=50)
            responses.append(tokenizer.decode(outputs[0]))

        # 计算奖励（网页4的奖励标准化处理）
        reward_inputs = tokenizer(responses, padding=True, return_tensors="pt").to("cuda")
        rewards = reward_model(**reward_inputs).logits

        # PPO优化步骤（网页5的GAE实现）
        stats = ppo_trainer.step(
            queries=prompts,
            responses=responses,
            rewards=rewards
        )
        print(f"PPO Step Stats: {stats}")

    sft_model.save_pretrained("./ppo_final_model")

```

```python
# 执行完整流程
if __name__ == "__main__":
    RewardModelTrainer().train()  # 阶段2
    ppo_training()                # 阶段3
```
